```python
# 1.

# taking input 
sentence =input("enter sentence:")
# split into words
words=sentence.split()
unique_words=set(words)

#print statements
print("unique words count:",len(unique_words))
print("unique words:", unique_words)
```

```python
# 2.

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# 3.

numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]

# Loop through list to compare values
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
        
print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
# 4.

prices = [450, 1200, 899, 1500, 300]
count = 0

# Check each price
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# 5.

attendance = ["P", "P", "A", "P", "P"]
# Count present days
present_days = attendance.count("P")
total_days = len(attendance)

# Percentage calculation
percentage = (present_days/total_days) * 100

print("Attendance Percentage:", percentage)
```
